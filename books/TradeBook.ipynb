```python
#METHOD_06
def Method_06(name, start, stop, targetDown):
    '''
    targetDown: Biên độ giảm
    '''
    
    desc = "Phương pháp 06: Mua ở giá giảm khi đạt mục tiêu biến động cao chiết khấu sâu"
    desc = desc + "\nBiên độ giảm: " + str(targetDown) + " (%)"

    capital = 100000000 #capital: Tiền vốn (ban đầu) 100.000.000

    stockLabel = name
    book = TradeBook(name=stockLabel)
    stock = StockObj(name=stockLabel)
    stock.Prepare()
    prices = stock.PRICES[start:stop]
    vols = stock.VOLS[start:stop]
    margins = stock.df_data['%'][start:stop]
    
    vol = 100

    
    desc = desc + "\nCổ phiếu: " + stockLabel 
    desc = desc + stock.StrSummary
    desc = desc + "\nDữ liệu: " + str(stop) + " về trước tính từ phiên thứ " + str(start) + " gần đây nhất" 
    print(desc)
    for i in range(start,stop):
        vol_i = vols[i]
        price_i = prices[i]
        margin_i = margins[i] #So với phiên trước
        
        #Điều kiện mua        
        if(margin_i <= targetDown) :
        #if(vol_i <= 0.5*np.mean(vols)) and (margin_i < -3) :
            print("Mua tại phiên thứ: " + str(i) + " với giá: " + str(price_i))
            book.Buy(price=price_i,vol=vol_i)

    book.PrintBook()

Method_06(name="FRT",start=0, stop=20, targetDown=-5)
Method_06(name="HBC",start=0, stop=20, targetDown=-6)
```

**PHƯƠNG PHÁP 07: MUA Ở GIÁ THẤP NHẤT TRONG NGÀY**

```python
#METHOD_07: MUA Ở GIÁ THẤP NHẤT TRONG NGÀY (với điều kiện giá giảm)
def Method_07(name, start, stop):
    '''
    targetDown: Biên độ giảm
    '''
    
    desc = "Phương pháp 06: Mua ở giá thấp nhất trong ngày"    

    capital = 100000000 #capital: Tiền vốn (ban đầu) 100.000.000

    stockLabel = name
    book = TradeBook(name=stockLabel)
    stock = StockObj(name=stockLabel)
    stock.Prepare()
    prices = stock.PRICES[start:stop]
    vols = stock.VOLS[start:stop]
    margins = stock.df_data['%'][start:stop]
    
    vol = 100

    
    desc = desc + "\nCổ phiếu: "
    desc = desc + stock.StrSummary
    desc = desc + "\nDữ liệu: " + str(stop) + " về trước tính từ phiên thứ " + str(start) + " gần đây nhất" 
    print(desc)
    for i in range(start,stop):
        vol_i = vols[i]
        price_i = prices[i]
        margin_i = margins[i] #So với phiên trước
        
        #Điều kiện mua        
        if(stock.df_data['Low'][i] == stock.df_data['Close'][i]) and (margin_i < -3):
        #if(vol_i <= 0.5*np.mean(vols)) and (margin_i < -3) :
            print("Mua tại phiên thứ: " + str(i) + " với giá: " + str(price_i))
            p_buy = stock.df_data['Low'][i]
            book.Buy(price=p_buy,vol=vol_i)

    book.PrintBook()

stocks = ["KBC","SCR","HBC","VND","HAX","MWG"]
for item in stocks:
    Method_07(name=item,start=0, stop=20)
```
